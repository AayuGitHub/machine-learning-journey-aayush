```python
import csv
from collections import defaultdict

def read_expenses(filename):
    expenses = defaultdict(float) # Changed to float for summing

    with open(filename, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            category = row["Category"]
            amount = float(row["Amount"])
            expenses[category] += amount
        return expenses

# Unindented these lines to make them run
expenses = read_expenses("expenses.csv")

# Print Summary by category
for category, total in expenses.items():
    print(f"Total spent on {category}: ${total:.2f}")
```

```python
import matplotlib.pyplot as plt

categories = list(expenses.keys())
totals = list(expenses.values())

plt.bar(categories, totals)
plt.xlabel("Category")
plt.ylabel("Total Amount ($)")
plt.title("Personal Expenses Summary")
plt.show()
```
